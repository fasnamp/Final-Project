### Data Cleaning

```python
df['Sleep Disorder'].isnull().sum()
```

```python
df['Sleep Disorder'].fillna("No Disorder",inplace=True)
```

```python
df
```

```python
df.columns=df.columns.str.replace(" ","_")
df.columns
```

```python
df
```

### Average Sleep Duration by Gender

```python
Sleep_by_gender=df.groupby("Gender")["Sleep_Duration"].mean()
Sleep_by_gender
```

### Average Stress Level by Occupation

```python
Stress_by_job=df.groupby("Occupation")["Stress_Level"].mean()
Stress_by_job
```

### Relationship Between Physical Activity And Sleep

```python
Activity_Sleep=df.groupby("Physical_Activity_Level")["Sleep_Duration"].mean()
Activity_Sleep
```

### Most Common BMI Category

```python
df["BMI_Category"].value_counts()
```

### Sleep Disorder Analysis

```python
df["Sleep_Disorder"].value_counts()
```

### Age Group Analysis


```python
df["Age_Group"]=pd.cut(df["Age"],bins=[20,30,40,50,60],labels=["20-30","30-40","40-50","50-60"])
age_sleep=df.groupby("Age_Group")["Sleep_Duration"].mean()
age_sleep
```

### Average sleep duration for each gender.

```python
avg_sleep_gender = df.groupby('Gender')['Sleep_Duration'].mean()
avg_sleep_gender
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,5))
plt.pie(Sleep_by_gender.values,labels=Sleep_by_gender.index,autopct='%1.1f%%',colors=['pink','blue'])
plt.title("Average Sleep Duration By Gender",color='black',fontsize=20,fontweight='bold')
plt.show()
```

```python
sleep_occupation=df.groupby("Occupation")["Sleep_Duration"].mean()
sleep_occupation
```

```python
plt.figure(figsize=(10,5))
plt.bar(sleep_occupation.index,sleep_occupation.values,color='grey')
plt.title("Average Sleep Duration By Occupation ",color='green',fontsize=20,fontweight='bold')
plt.xlabel("Occupation",color='black',fontsize=16,fontweight='bold')
plt.ylabel("Sleep Hours",color='black',fontsize=16,fontweight='bold')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x="Occupation",y="Stress_Level",data=df,palette="viridis")
plt.title("Stress Level by Occupation", fontsize=20, fontweight='bold')
plt.xlabel("Occupation",color='black',fontsize=16,fontweight='bold')
plt.ylabel("Stress level",color='black',fontsize=16,fontweight='bold')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x="Age",y="Sleep_Duration",data=df)
plt.title("Age vs Sleep Duration",fontsize=20, fontweight='bold')
plt.xlabel("Age",color='black',fontsize=16,fontweight='bold')
plt.ylabel("Sleep Duration",color='black',fontsize=16,fontweight='bold')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='Gender', y='Sleep_Duration', data=avg_sleep_gender, palette="pastel")
plt.title("Average Sleep Duration by Gender",fontsize=20, fontweight='bold')
plt.xlabel("Gender")
plt.ylabel("Average Sleep Duration (hours)")
plt.show()
```

```python
quality_BMI=df.groupby("BMI_Category")["Quality_of_Sleep"].mean()
```

```python
plt.figure(figsize=(10,5))
plt.barh(quality_BMI.index,quality_BMI.values,color='green')
plt.title("Sleep Quality by BMI Category",fontsize=20, fontweight='bold')
plt.xlabel("Quality Sleep",color='black',fontsize=16,fontweight='bold')
plt.ylabel("BMI Category",color='black',fontsize=16,fontweight='bold')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot( x="Age", y="Systolic_BP",data=df, marker="o", color="dodgerblue")
plt.title("Average Systolic BP by Age", fontsize=20, fontweight='bold')
plt.ylabel("Systolic BP", fontsize=16, fontweight='bold')
plt.xlabel("Age", fontsize=16, fontweight='bold')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(x="Physical_Activity_Level", y="Quality_of_Sleep", data=df)
plt.title("Physical Activity vs Sleep Quality", fontsize=20, fontweight='bold')
plt.ylabel("Sleep Quality", fontsize=16, fontweight='bold')
plt.xlabel("Physical Activity", fontsize=16, fontweight='bold')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=["int64", "float64"])

plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True)
plt.title("Correlation Heatmap",fontsize=20, fontweight='bold')
plt.show()
```

```python
stress_by_steps = df.groupby("Daily_Steps")["Stress_Level"].mean().reset_index()
plt.figure(figsize=(10,6))
plt.plot(stress_by_steps["Daily_Steps"], stress_by_steps["Stress_Level"], marker="o", color="crimson")
plt.title("Daily Steps vs Stress Level", fontsize=20, fontweight='bold')
plt.ylabel("Stress Level", fontsize=16, fontweight='bold')
plt.xlabel("Daily Steps", fontsize=16, fontweight='bold')
plt.show()
```

```python
sleep_by_occupation = df.groupby("Occupation")["Sleep_Disorder"].value_counts().unstack(fill_value=0)
plt.figure(figsize=(12,6))
sleep_by_occupation.plot(kind="bar", stacked=True, colormap="Set2")
plt.title("Sleep Disorder Count by Occupation", fontsize=20, fontweight="bold")
plt.ylabel("Count", fontsize=16, fontweight="bold")
plt.xlabel("Occupation", fontsize=16, fontweight="bold")
plt.xticks(rotation=45, ha="right")
plt.legend(title="Sleep Disorder", fontsize=12, title_fontsize=14)
plt.tight_layout()
plt.show()
```
